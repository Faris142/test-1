```python
def add_expense():
    expense = {}
    expense['amount'] = float(input("Enter the expense amount: "))
    expense['category'] = input("Enter the expense category: ")
    expense['date'] = input("Enter the expense date (YYYY-MM-DD): ")
    return expense

def add_income():
    income = {}
    income['amount'] = float(input("Enter the income amount: "))
    income['category'] = input("Enter the income category: ")
    income['date'] = input("Enter the income date (YYYY-MM-DD): ")
    return income

def display_transactions(transactions):
    if not transactions:
        print("No transactions found.")
    else:
        for idx, transaction in enumerate(transactions):
            print(f"{idx+1}. Amount: {transaction['amount']}, Category: {transaction['category']}, Date: {transaction['date']}")

def update_transaction(transactions):
    display_transactions(transactions)
    transaction_index = int(input("Enter the index of the transaction you want to update: ")) - 1

    if transaction_index < 0 or transaction_index >= len(transactions):
        print("Invalid transaction index.")
    else:
        transaction = transactions[transaction_index]
        print("\nCurrent transaction details:")
        print(f"Amount: {transaction['amount']}, Category: {transaction['category']}, Date: {transaction['date']}")

        transaction['amount'] = float(input("Enter the updated amount: "))
        transaction['category'] = input("Enter the updated category: ")
        transaction['date'] = input("Enter the updated date (YYYY-MM-DD): ")

        print("Transaction updated successfully.")

def delete_transaction(transactions):
    display_transactions(transactions)
    transaction_index = int(input("Enter the index of the transaction you want to delete: ")) - 1

    if transaction_index < 0 or transaction_index >= len(transactions):
        print("Invalid transaction index.")
    else:
        transaction = transactions.pop(transaction_index)
        print("Transaction deleted successfully.")
        print(f"Deleted transaction details - Amount: {transaction['amount']}, Category: {transaction['category']}, Date: {transaction['date']}")
```

```python
expenses = {}
incomes = {}

while True:
    print("\n----- Expense and Income Management Menu-----")
    print("1. Add an expense")
    print("2. Add an income")
    print("3. View transactions")
    print("4. Update a transaction")
    print("5. Delete a transaction")
    print("6. Generate reports")
    print("0. Exit")

    choice = input("Enter your choice: ")

    if choice == '1':
        expense = add_expense()
        category = expense['category']
        if category not in expenses:
            expenses[category] = []
        expenses[category].append(expense)
        print("Expense added successfully!")

    elif choice == '2':
        income = add_income()
        category = income['category']
        if category not in incomes:
            incomes[category] = []
        incomes[category].append(income)
        print("Income added successfully!")

    elif choice == '3':
        print("\n----- View Transactions -----")
        print("1. Sort by date")
        print("2. Sort by category")
        sort_choice = input("Enter your choice: ")

        if sort_choice == '1':
            print("\n----- Sorted by Date -----")
            for category, transaction_list in expenses.items():
                print(f"Category: {category}")
                transaction_list.sort(key=lambda x: x['date'])
                display_transactions(transaction_list)
            for category, transaction_list in incomes.items():
                print(f"Category: {category}")
                transaction_list.sort(key=lambda x: x['date'])
                display_transactions(transaction_list)

        elif sort_choice == '2':
            print("\n----- Sorted by Category -----")
            sorted_expenses = sorted(expenses.items())
            sorted_incomes = sorted(incomes.items())
            for category, transaction_list in sorted_expenses:
                print(f"Category: {category}")
                display_transactions(transaction_list)
            for category, transaction_list in sorted_incomes:
                print(f"Category: {category}")
                display_transactions(transaction_list)
        else:
            print("Invalid choice.")

    elif choice == '4':
        print("\n----- Update a Transaction -----")
        print("1. Update an expense")
        print("2. Update an income")
        update_choice = input("Enter your choice: ")

        if update_choice == '1':
            category = input("Enter the category of the expense: ")
            if category in expenses:
                update_transaction(expenses[category])
            else:
                print("No expenses found for the given category.")
        elif update_choice == '2':
            category = input("Enter the category of the income: ")
            if category in incomes:
              update_transaction(incomes[category])
            else:
              print("Invalid choice.")

    elif choice == '5':
        print("\n----- Delete a Transaction -----")
        print("1. Delete an expense")
        print("2. Delete an income")
        delete_choice = input("Enter your choice: ")

        if delete_choice == '1':
            category = input("Enter the category of the expense: ")
            if category in expenses:
                delete_transaction(expenses[category])
            else:
                print("No expenses found for the given category.")
        elif delete_choice == '2':
            category = input("Enter the category of the income: ")
            if category in incomes:
                delete_transaction(incomes[category])
            else:
                print("No incomes found for the given category.")
        else:
            print("Invalid choice.")
    elif choice == '6':
        print("\n----- Generate Reports -----")
        print("1. Total expenses")
        print("2. Total income")
        print("3. Spending by category")
        print("4. Transactions in each category")
        print("5. Average transaction value by category")
        report_choice = input("Enter your choice: ")

        if report_choice == '1':
            total_expenses = sum([sum([expense['amount'] for expense in transaction_list]) for transaction_list in expenses.values()])
            print(f"Total expenses: {total_expenses}")
        elif report_choice == '2':
            total_income = sum([sum([income['amount'] for income in transaction_list]) for transaction_list in incomes.values()])
            print(f"Total income: {total_income}")
        elif report_choice == '3':
            print("\n----- Spending by Category -----")
            for category, transaction_list in expenses.items():
                total_spending = sum([expense['amount'] for expense in transaction_list])
                print(f"Category: {category}, Total spending: {total_spending}")
        elif report_choice == '4':
            print("\n----- Transactions in Each Category -----")
            for category, transaction_list in expenses.items():
                print(f"Category: {category}")
                display_transactions(transaction_list)
            for category, transaction_list in incomes.items():
                print(f"Category: {category}")
                display_transactions(transaction_list)
        elif report_choice == '5':
            print("\n----- Average Transaction Value by Category -----")
            for category, transaction_list in expenses.items():
                total_spending = sum([expense['amount'] for expense in transaction_list])
                average_spending = total_spending / len(transaction_list)
                print(f"Category: {category}, Average spending: {average_spending}")
        else:
            print("Invalid choice.")
    elif choice == '0':
        print("Thank you for using the Expense and Income Management System!")
        break
    else:
        print("Invalid choice.")
```
